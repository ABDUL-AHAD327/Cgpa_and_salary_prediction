THIS IS A MODEL THAT SHOWS THE RELATIONSHIP OF CGPA ON SALARIES(LPA)

```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r"C:\Users\hp\OneDrive\Desktop\VScode\Machine_Learning\cgpa_salarydata\placement.csv")

print(df.head())

print(df.shape) #(num_rows, num_columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size
plt.figure(figsize=(8, 5))

# Scatter plot of CGPA vs Salary Package
sns.scatterplot(x=df['cgpa'], y=df['package'], color='blue', alpha=0.7)

# Labels and title
plt.xlabel("CGPA")
plt.ylabel("Salary Package (LPA)")
plt.title("CGPA vs Salary Package")
plt.grid(True)

plt.show()
```

WE CAN NOW SEE THE PLOT OF OUR DATASET WITH CGPA ON THE X-AXIS AND THE SALARIES ON THE Y-AXIS

SINCE THE NUMBER OF FEATURES ARE TWO THE HYPOTHESIS WILL BE θ(0)+θ(1)X(1)

```python
import numpy as np


# Load dataset
X = df["cgpa"].values  # Independent variable
y = df["package"].values  # Dependent variable
m = len(y)  # Number of data points

# Initialize parameters
theta_0 = 0  # Intercept
theta_1 = 0  # Slope
alpha = 0.01  # Learning rate
iterations = 10  # Number of iterations
cost_history = []  # To store cost function values

# Gradient Descent Algorithm
for i in range(iterations):
    # Predictions
    h = theta_0 + theta_1 * X

    # Compute cost function
    cost = (1 / (2 * m)) * np.sum((h - y) ** 2)
    cost_history.append(cost)

    # Compute gradients
    d_theta_0 = (1 / m) * np.sum(h - y)
    d_theta_1 = (1 / m) * np.sum((h - y) * X)

    # Update parameters
    theta_0 -= alpha * d_theta_0
    theta_1 -= alpha * d_theta_1

    print(f"Iteration {i+1}: Cost = {cost:.4f}, theta_0 = {theta_0:.4f}, theta_1 = {theta_1:.4f}")

# Plot Cost Function Over Iterations
plt.plot(range(1, iterations + 1), cost_history, marker='o', linestyle='-')
plt.xlabel("Iteration")
plt.ylabel("Cost Function (J)")
plt.title("Cost Function Over Iterations")
plt.grid(True)
plt.show()
```

HERE WE CAN SEE THAT THE THE COST DECREASES WITH EACH ITERARION.YOU CAN TWEAK THIS BY CHANGING THE VALUE OF LR OR THE NUMBER OF ITERATIONS

```python
theta_0_opt = 0.0568
theta_1_opt = 0.4234 #as found above using gradient descent

# Compute the hypothesis
X_line = np.linspace(min(X), max(X), 100)  # Generate X values for a smooth line
y_pred = theta_0_opt + theta_1_opt * X_line  # h(x) = theta_0 + theta_1 * X

# Plot data points
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', alpha=0.7, label="Actual Data")  # Scatter plot of actual data
plt.plot(X_line, y_pred, color='red', linewidth=2, label="Hypothesis (Regression Line)")  # Regression line


plt.xlabel("CGPA")
plt.ylabel("Salary Package (LPA)")
plt.title("Linear Regression: CGPA vs Salary")
plt.legend()
plt.grid(True)
plt.show()
```

THE ACHIEVED LINE IS GOOD BUT NOT THE BEST. LETS TRY GRADIENT DESCENT WITH A 1000 ITERATIONS AND LR 0.001

```python
import numpy as np


# Load dataset
X = df["cgpa"].values  # Independent variable
y = df["package"].values  # Dependent variable
m = len(y)  # Number of data points

# Initialize parameters
theta_0 = 0  # Intercept
theta_1 = 0  # Slope
alpha = 0.0001  # Learning rate
iterations = 1000  # Number of iterations
cost_history = []  # To store cost function values

# Gradient Descent Algorithm
for i in range(iterations):
    # Predictions
    h = theta_0 + theta_1 * X

    # Compute cost function
    cost = (1 / (2 * m)) * np.sum((h - y) ** 2)
    cost_history.append(cost)

    # Compute gradients
    d_theta_0 = (1 / m) * np.sum(h - y)
    d_theta_1 = (1 / m) * np.sum((h - y) * X)

    # Update parameters
    theta_0 -= alpha * d_theta_0
    theta_1 -= alpha * d_theta_1

    print(f"Iteration {i+1}: Cost = {cost:.4f}, theta_0 = {theta_0:.4f}, theta_1 = {theta_1:.4f}")

# Plot Cost Function Over Iterations
plt.plot(range(1, iterations + 1), cost_history, marker='o', linestyle='-')
plt.xlabel("Iteration")
plt.ylabel("Cost Function (J)")
plt.title("Cost Function Over Iterations")
plt.grid(True)
plt.show()
```

TRYING TO FIND THE NEW HYPOTHESIS USING UPDATED THETAS 

```python
theta_0_opt = 0.0565
theta_1_opt = 0.4212 #as found above using gradient descent

# Compute the hypothesis
X_line = np.linspace(min(X), max(X), 100)  # Generate X values for a smooth line
y_pred = theta_0_opt + theta_1_opt * X_line  # h(x) = theta_0 + theta_1 * X

# Plot data points
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', alpha=0.7, label="Actual Data")  # Scatter plot of actual data
plt.plot(X_line, y_pred, color='red', linewidth=2, label="Hypothesis (Regression Line)")  # Regression line


plt.xlabel("CGPA")
plt.ylabel("Salary Package (LPA)")
plt.title("Linear Regression: CGPA vs Salary")
plt.legend()
plt.grid(True)
plt.show()
```

EVEN AFTER USING A LOWER LR AND HIGH NUMBER OF ITERATIONS THE LINE IS NOT PERFECT.
LETS TRY USING SCKIT LEARN FOR THIS

```python
from sklearn.linear_model import LinearRegression

# Reshape X for sklearn
X_reshaped = X.reshape(-1, 1)

# Train Linear Regression Model
model = LinearRegression()
model.fit(X_reshaped, y)

# Get optimized parameters
theta_0_sklearn = model.intercept_
theta_1_sklearn = model.coef_[0]
print(f"Scikit-Learn Optimized theta_0: {theta_0_sklearn:.4f}, theta_1: {theta_1_sklearn:.4f}")

# Make Predictions
y_pred_sklearn = model.predict(X_reshaped)

# Plot Regression Line
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', alpha=0.7, label="Actual Data")
plt.plot(X, y_pred_sklearn, color='green', linewidth=2, label="Scikit-Learn Regression Line")
plt.xlabel("CGPA")
plt.ylabel("Salary Package (LPA)")
plt.title("Scikit-Learn Linear Regression: CGPA vs Salary")
plt.legend()
plt.grid(True)
plt.show()
```

WE CAN SEE THIS IS MUCH BETTER BECAUSE SCKIT LEARN AUTOMATICALLY CHOOSES THETA 0 AND THETA 1 USING LEAST SQUARED METHOD.

NOW THE LINE LOOKS OPTIMIZED, TIME TO TEST IT

```python
import numpy as np

# Enter new CGPA values to test
new_cgpa = np.array([2.0,8.25, 9.8]).reshape(-1, 1)  # Example values

# Predict salary packages for the new CGPA values
predicted_salaries = model.predict(new_cgpa)

# Display results
for cgpa, salary in zip(new_cgpa.flatten(), predicted_salaries):
    print(f"Predicted Salary for CGPA {cgpa}: {salary:.2f} LPA")
```
